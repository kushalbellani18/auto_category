# Data Preprocessing

```python
df_ = df[df["org_code"].isin(org_code_possible_lists)]
df_ = df_[["_id", "subject_content_text"]]

df_.sample(n=9)
```

```python
import string

def clean_text(txt):
    txt = txt.lower()
    clean_txt = ""
    
    for t in txt:
        if t not in string.punctuation:
            clean_txt += t
            
    return clean_txt
```

```python
df_["subject_content_text_cleaned"] = df_["subject_content_text"].progress_apply(clean_text)
```

# Word Embeddings

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
df_["minilm_embeddings_cleaned"] = df_["subject_content_text_cleaned"].progress_apply(
    lambda row: model.encode(row, convert_to_tensor=True).tolist()
)
```
